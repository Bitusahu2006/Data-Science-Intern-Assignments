```python
#  Broadcasting
# Create two arrays:
a = np.array([1,2,3])
b = np.array([1,2,3])

print("a+b =", a+b)
print("a-b =", a-b)
print("a*b =", a*b)
print("a/b =", a/b)
```

```python
# Given scores = [72, 85, 91, 68, 74, 88, 62, 95, 77,
# 80], compute: mean, median (np.median), std,
# min, max, and range (max-min).
```

```python
print(np.max(arr)-np.min(arr))
```

```python
# For a 3×4 random integer array
# (np.random.randint(1,100,size=(3,4))): compute
# sum across all elements, row-wise sum (axis=1),
# and column-wise mean (axis=0).
```

```python
arr = np.random.randint(1,100,size=(3,4))
print(arr)
```

```python
# Row-wise Sum (axis=1)
print(np.sum(arr, axis=1))
```

```python
# Column-wise Mean (axis=0)
print(np.mean(arr, axis=0))
```
